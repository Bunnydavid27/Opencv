```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("C:/Users/david/Data Sciences/ML/PROJECT ML/Images/maria sharapova.jpg")
plt.imshow(img)
```

```python
mai_scale = cv2.Canny(grey_scale,80,200)
plt.imshow(mai_scale)
```
