```python
import cv2
import matplotlib.pyplot as plt
```

```python
cap= cv2.VideoCapture(0)
while 1:
    ret,img = cap.read()
    
    imgi = img[:, :, 2]
    cv2.imshow('img',imgi)
    k = cv2.waitKey(30) & 0xff
    if k==ord('q'):
        break
        
cap.release()
cv2.waitKey(0)
    
```
